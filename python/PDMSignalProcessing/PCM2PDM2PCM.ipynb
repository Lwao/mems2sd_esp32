```python
import numpy as np
import soundfile as sf
import sounddevice as sd
import matplotlib.pyplot as plt
%matplotlib inline

from PDM2PCM import *
from PCM2PDM import *
```

```python
t_end = 1
fs = int(1e6)
os = 32

t = np.arange(start=0, stop=t_end, step=1/fs)
demo_data = np.sin(2*np.pi * 2000*t) + np.random.randn(len(t))/10

pcm2pdm = PCM2PDM(demo_data, fs_pcm=fs, os=1)
pdm2pcm = PDM2PCM(pcm2pdm.data_pdm, fs_pdm=fs, os=os, stages=2, delay=1)

pcm2pdm.mag_plot(short=True)
pdm2pcm.mag_plot(short=True)
```
